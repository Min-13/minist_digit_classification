1. Import TensorFlow



```python
# Import TensorFlow for building and training the neural network
import tensorflow as tf
```

2. Load MINIST Dataset

```python
# Load the MNIST dataset (handwritten digits 0–9)
# It automatically splits into training and testing sets
mnist = tf.keras.datasets.mnist

# X = images (28x28 grayscale), y = labels (0–9)
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

3. Inspect Dataset Shape

```python
# Check training image shape
# Expected: (60000, 28, 28)
X_train.shape

# Check training label shape
# Expected: (60000,)
y_train.shape
```

4. View a Sample Image (Raw Pixel Values)

```python
# Inspect first image (28x28 matrix of pixel values from 0–255)
X_train[0]
```

5. Normalize Data

```python
# Normalize pixel values from range [0, 255] to [0, 1]
# This improves training stability and convergence
X_train, X_test = X_train / 255.0, X_test / 255.0
```

6. Build the Neural Network

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Input, Dropout

# Create a Sequential neural network model
model = Sequential()

# Define input shape (28x28 grayscale image)
model.add(Input(shape=(28, 28)))

# Flatten image into 784-dimensional vector
model.add(Flatten())

# Hidden layer with 128 neurons and ReLU activation
model.add(Dense(128, activation='relu'))

# Dropout layer to reduce overfitting (randomly disables 20% of neurons)
model.add(Dropout(0.2))

# Output layer with 10 neurons (one per digit class)
# No activation because we use logits (softmax applied later)
model.add(Dense(10))
```

7. Test Untrained Model (Sanity Check)

```python
# Generate prediction before training (random weights)
nonsense_prediction = model(X_train[:1]).numpy()

nonsense_prediction
```

8. Convert Logits to Probabilites

```python
# Apply softmax to convert logits to probability distribution
tf.nn.softmax(nonsense_prediction).numpy().sum()
```

9. Define Loss Function

```python
from tensorflow.keras.losses import SparseCategoricalCrossentropy

# Use sparse categorical crossentropy
# from_logits=True because output layer does NOT use softmax
loss_fn = SparseCategoricalCrossentropy(from_logits=True)

# Compute initial loss (before training)
loss_fn(y_train[:1], nonsense_prediction)
```

10. Compile Model

```python
from tensorflow.keras.optimizers import Adam

# Use Adam optimizer with learning rate 0.01
optimizer = Adam(learning_rate=0.01)

# Compile model with optimizer, loss, and accuracy metric
model.compile(
    optimizer=optimizer,
    loss=loss_fn,
    metrics=['accuracy']
)
```

11. Train Model

```python
# Train for 10 epochs on training dataset
model.fit(X_train, y_train, epochs=10)
```

12. Evaluate Model

```python
# Evaluate performance on unseen test data
model.evaluate(X_test, y_test)
```

13. Make Final Prediction

```python
# Convert model output to probabilities
tf.nn.softmax(model(X_test[:1])).numpy()
```
